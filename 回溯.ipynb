回溯法 图示理解：
本质是一个深度优先搜索过程，

```python
class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:
        path=[]
        res=[]
        def backtracking(start_index):
            print(f'递归start_ind: {start_index}')

            # if (len(path)==k):
            #     res.append(path[:])  # 注意是path 的当前状态的副本
            #     return
            
            for i in range(start_index, n+1):
                # 遍历【1,2,3..,n】
                path.append(i)
                print(path,'<-----')
                backtracking(i+1)
                path.pop()
                print(path,'----->')

        backtracking(1)
        return res
完整的搜过程：
递归start_ind: 1
[1] <-----
递归start_ind: 2
[1, 2] <-----
递归start_ind: 3
[1, 2, 3] <-----
递归start_ind: 4
[1, 2, 3, 4] <-----
递归start_ind: 5
[1, 2, 3] ----->
[1, 2] ----->
[1, 2, 4] <-----
递归start_ind: 5
[1, 2] ----->
[1] ----->
[1, 3] <-----
递归start_ind: 4
[1, 3, 4] <-----
递归start_ind: 5
[1, 3] ----->
[1] ----->
[1, 4] <-----
递归start_ind: 5
[1] ----->
[] ----->
[2] <-----
递归start_ind: 3
[2, 3] <-----
递归start_ind: 4
[2, 3, 4] <-----
递归start_ind: 5
[2, 3] ----->
[2] ----->
[2, 4] <-----
递归start_ind: 5
[2] ----->
[] ----->
[3] <-----
递归start_ind: 4
[3, 4] <-----
递归start_ind: 5
[3] ----->
[] ----->
[4] <-----
递归start_ind: 5
[] ----->
```
